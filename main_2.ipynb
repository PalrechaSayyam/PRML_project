Results Across All Models (All Tasks)

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

SAVE_DIR = "results/plots"

baseline = pd.read_csv("results/tox21_baseline_results.csv")
optimized = pd.read_csv("results/tox21_optimized_results_new.csv")
graphconv = pd.read_csv("results/tox21_gnn_results_graphconv.csv")
gat = pd.read_csv("results/tox21_gnn_results_GAT.csv")

baseline_filtered = baseline[["Task", "Model", "ROC-AUC", "PR-AUC"]].copy()
baseline_filtered["Model"] = baseline_filtered["Model"].replace({"Random Forest": "RF"})
optimized_filtered = optimized[["Task", "Model", "ROC-AUC", "PR-AUC"]].copy()
graphconv_filtered = graphconv[graphconv["Split"] == "TEST"][["Task", "Model", "ROC-AUC", "PR-AUC"]].copy()
gat_filtered = gat[gat["Split"] == "TEST"][["Task", "Model", "ROC-AUC", "PR-AUC"]].copy()

summary_df = pd.concat([
    baseline_filtered,
    optimized_filtered,
    graphconv_filtered,
    gat_filtered
], ignore_index=True)

tasks = sorted(summary_df["Task"].unique())
summary_df["Task"] = pd.Categorical(summary_df["Task"], categories=tasks, ordered=True)
models_to_plot = ["RF", "SVM", "RF_Optimized", "XGB_Optimized", "GraphConv", "GAT"]

def plot_metric(metric_name, output_filename):
    plt.figure(figsize=(10, 6))
    for model in models_to_plot:
        model_df = summary_df[summary_df["Model"] == model].sort_values(by="Task")
        if not model_df.empty:
            plt.plot(model_df["Task"], model_df[metric_name], marker="o", linewidth=2, label=model)

    plt.xlabel("Tasks", fontsize=12)
    plt.ylabel(metric_name, fontsize=12)
    plt.title(f"{metric_name} across Tox21 tasks", fontsize=14)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(SAVE_DIR, output_filename), dpi=300)
    # plt.show()

plot_metric("ROC-AUC", "roc_auc_across_tasks.png")
plot_metric("PR-AUC", "pr_auc_across_tasks.png")
```

Baseline Interpretation

```python
import pandas as pd

baseline_df = pd.read_csv("results/tox21_baseline_results.csv")
baseline_df = baseline_df[["Task", "Model", "ROC-AUC", "PR-AUC"]]
tasks = baseline_df["Task"].unique()

# Initialize lists for comparison
better_roc = []
better_pr = []

for task in tasks:
    subset = baseline_df[baseline_df["Task"] == task]

    rf_row = subset[subset["Model"].str.contains("Random", case=False)]
    svm_row = subset[subset["Model"].str.contains("SVM", case=False)]

    if not rf_row.empty and not svm_row.empty:
        rf_roc, rf_pr = rf_row.iloc[0]["ROC-AUC"], rf_row.iloc[0]["PR-AUC"]
        svm_roc, svm_pr = svm_row.iloc[0]["ROC-AUC"], svm_row.iloc[0]["PR-AUC"]

        roc_better = "RF" if rf_roc > svm_roc else "SVM"
        pr_better = "RF" if rf_pr > svm_pr else "SVM"

        better_roc.append((task, roc_better))
        better_pr.append((task, pr_better))

roc_df = pd.DataFrame(better_roc, columns=["Task", "Better_ROC_Model"])
pr_df = pd.DataFrame(better_pr, columns=["Task", "Better_PR_Model"])

comparison_df = pd.merge(roc_df, pr_df, on="Task")
print("\n=== Model Comparison per Task ===")
print(comparison_df.to_string(index=False))

rf_better_tasks = comparison_df[
    (comparison_df["Better_ROC_Model"] == "RF") | (comparison_df["Better_PR_Model"] == "RF")
]["Task"].unique()

svm_better_tasks = comparison_df[
    (comparison_df["Better_ROC_Model"] == "SVM") | (comparison_df["Better_PR_Model"] == "SVM")
]["Task"].unique()

summary = f"""
Baseline Model Comparison Summary
---------------------------------
Total tasks analyzed: {len(tasks)}

RF performed better (ROC-AUC and/or PR-AUC) on: {', '.join(rf_better_tasks)}
SVM performed better (ROC-AUC and/or PR-AUC) on: {', '.join(svm_better_tasks)}

Observations:
- Tasks where RF performs better likely have non-linear relationships or substructure-level signal
  that RF's ensemble of trees can capture, even with some noise or imbalance.
- Tasks where SVM performs better are probably more linearly separable in the ECFP fingerprint space,
  suggesting that the active vs inactive molecules are well separated by a simple hyperplane.
- SVM may also dominate when data imbalance is less severe, since it does not inherently rebalance samples.
- Conversely, RF tends to perform more robustly when the data is noisy or mildly imbalanced, due to bagging and random feature selection.

Interpretation:
These performance splits imply that some Tox21 endpoints (e.g., receptor-based assays with consistent
chemical scaffolds) are easier to classify linearly, while others (e.g., stress-response assays) need
nonlinear models like RF that can capture higher-order feature interactions.

Next Step:
This justifies tuning RF/XGB (to improve imbalance handling) and later comparing to GNN models
to see if structure-based learning helps in tasks that defeated both SVM and RF.
"""

# Save summary
with open("docs/baseline_interpretation.txt", "w") as f:
    f.write(summary)

print("\nInterpretive summary saved at: docs/step1_baseline_interpretation.txt")
```

Extract Best Hyperparameters from optimized pipeline

```python
import os
import joblib
import pandas as pd

MODEL_DIR = "results/models"
OUTPUT_CSV = "results/optimized_hyperparameters.csv"

models = [f for f in os.listdir(MODEL_DIR) if f.endswith("_optimized.pkl")]

records = []
for fname in models:
    path = os.path.join(MODEL_DIR, fname)
    model = joblib.load(path)

    model_name = "RF_Optimized" if "RF" in fname else "XGB_Optimized"
    task_name = fname.replace("_RF_optimized.pkl", "").replace("_XGB_optimized.pkl", "")

    if hasattr(model, "model") and hasattr(model.model, "get_params"):
        params = model.model.get_params()
    elif hasattr(model, "get_params"):
        params = model.get_params()
    else:
        print(f"Skipping {fname}  no get_params() method.")
        continue

    key_params = {k: v for k, v in params.items() if k in [
        "n_estimators", "max_depth", "min_samples_split", "min_samples_leaf",
        "max_features", "learning_rate", "subsample", "colsample_bytree"
    ]}

    records.append({"Task": task_name, "Model": model_name, **key_params})

optimized_df = pd.DataFrame(records)
optimized_df.to_csv(OUTPUT_CSV, index=False)
print(f"Extracted optimized parameters saved to: {OUTPUT_CSV}")
print(optimized_df.head(10))
```

Hybrid: GNN + RF

```python
from hybrid_gnn_rf import train_gnn_model
```

```python
import deepchem as dc
tasks, graph_sets, _ = dc.molnet.load_tox21(featurizer="GraphConv")
train_graph, valid_graph, test_graph = graph_sets

_, ecfp_sets, _ = dc.molnet.load_tox21(featurizer="ECFP")
train_ecfp, valid_ecfp, test_ecfp = ecfp_sets
```

```python
# --- Step 1: Train GNN ---
# We train *once* on the train set
gnn_model = train_gnn_model(tasks, train_graph, valid_graph)
```

```python
# from hybrid_gnn_rf import get_ecfp_fps
import numpy as np

# --- Step 2: Get *true* GNN embeddings ---
print("Extracting GNN embeddings...")
X_train_gnn = gnn_model.predict_embedding(train_graph)
X_valid_gnn = gnn_model.predict_embedding(valid_graph)
# X_test_gnn = gnn_model.predict_embedding(test_graph) # For final testing

# --- Step 3: Get ECFP features *for the same molecules* ---
print("Generating ECFP features for the *same* molecules...")
ecfp_featurizer = dc.feat.CircularFingerprint(size=1024, radius=2)

# We must iterate to find featurization failures
X_train_ecfp_list = []
X_valid_ecfp_list = []
train_indices_to_keep = []
valid_indices_to_keep = []

# Process training set
print("Robustly featurizing ECFP for train set...")
for i, smiles in enumerate(train_graph.ids):
    fp = ecfp_featurizer.featurize([smiles]) # Featurize one at a time
    if fp.shape[0] > 0: # Check if featurization was successful (fp[0] is not None)
        X_train_ecfp_list.append(fp[0])
        train_indices_to_keep.append(i)
    # else: print(f"ECFP failed for train SMILES: {smiles}") # Uncomment to see fails
        
# Process validation set
print("Robustly featurizing ECFP for valid set...")
for i, smiles in enumerate(valid_graph.ids):
    fp = ecfp_featurizer.featurize([smiles])
    if fp.shape[0] > 0:
        X_valid_ecfp_list.append(fp[0])
        valid_indices_to_keep.append(i)

# Convert lists to numpy arrays
X_train_ecfp = np.array(X_train_ecfp_list)
X_valid_ecfp = np.array(X_valid_ecfp_list)
# X_train_ecfp shape is now (6264, 1024)
```

```python
import numpy as np
# --- Step 4: Filter *all* datasets to the common, successful indices ---
print("Filtering all datasets to common successful molecules...")

# Filter GNN embeddings to match
X_train_gnn = X_train_gnn[train_indices_to_keep]
X_valid_gnn = X_valid_gnn[valid_indices_to_keep]
# X_train_gnn shape is now (6264, N)

# Filter labels and weights (y, w) to match
y_train = train_graph.y[train_indices_to_keep]
w_train = train_graph.w[train_indices_to_keep]

y_valid = valid_graph.y[valid_indices_to_keep]
w_valid = valid_graph.w[valid_indices_to_keep]

print(f"Filtered train GNN shape: {X_train_gnn.shape}")
print(f"Filtered train ECFP shape: {X_train_ecfp.shape}")
print(f"Filtered valid GNN shape: {X_valid_gnn.shape}")
print(f"Filtered valid ECFP shape: {X_valid_ecfp.shape}")
```

```python
from hybrid_gnn_rf import train_and_eval_rf
# --- Step 5: Build Hybrid Feature Set ---
# Now, the dimensions *must* match
print("Concatenating features...")
X_train_hybrid = np.concatenate([X_train_gnn, X_train_ecfp], axis=1)
X_valid_hybrid = np.concatenate([X_valid_gnn, X_valid_ecfp], axis=1)

print(f"Hybrid feature shape: {X_train_hybrid.shape}")
```

```python
# --- Step 6: Train and Evaluate RF ---
df_results = train_and_eval_rf(
    tasks,
    X_train_hybrid, y_train, w_train,
    X_valid_hybrid, y_valid, w_valid
)

df_results.to_csv("results/hybrid_gnn_rf_results.csv", index=False)
print("\nSaved results -> results/hybrid_gnn_rf_results.csv")
```
